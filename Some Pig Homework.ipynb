# Some Pig Simulation

A player has 7 turns. A turn consists of a number of rolls of a single die. The number of rolls is decided by the player as he goes. The turn can end at the player's choosing. At any time during a turn, a roll of a 1 ends the turn and the player's total score remains the same. Each roll > 1 during a turn is added to a turn total. When the player chooses to end the turn (not having rolled  a 1), the turn total is added to the total score. The objective is to get the total score to 100 in the 7 turns.

There will be a generic Player class. GP will roll once per turn, then stand(function) to end the turn. 

Two other players will be created that have different set playing strategies. One will stand when the turn total reaches 15, as 7 successful turns will result in a score of 100.

The third player will randomly stand after 3-7 rolls. This increases his chance at a high scoring turn while occassionally playing conservatively.

Time allowing, I will create a fourth player that will stand after rolling the same number back-to-back.

```python
import random
import math
import statistics as st
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
class Player():
    """
    Responsibilities:
    - rolling the die
    - choosing to roll or stand
    - keep track of turn total
    """
    def __init__(self):
        self.turn_total = 0
    
         
    def roll(self):
        return random.randint(1, 6)
    
    
    def stand(self):
        return self.turn_total
    
   
    def reset_turn_total(self):
        self.turn_total = 0
        return self.turn_total
    
    
    def play_and_get_turn_total(self):
        self.reset_turn_total()
        roll = self.roll()
        if roll > 1:
            self.turn_total += roll
            self.stand()
        elif roll == 1:
            return self.turn_total
        else:
            self.turn_total += self.roll()
        return self.turn_total
    
 
                    
```

```python
class Game():
    """
    Responsibilities:
    - keeping total score
    - ending turns with 1
    - ending the game at 7 turns
    """
    
    
    def __init__(self, player):
        
        self.player = Player()
        self.overall_score = 0
        self.turns = 0 
        
        
    def get_score_from_turn(self):
        self.overall_score += player.turn_total
        return self.overall_score
    
    
    def take_turn(self):
        player.play_and_get_turn_total()  
        self.turns += 1
        return self.turns, player.turn_total
    
    
    def play_game(self, player):
        while self.turns < 7:
            self.take_turn()
            self.get_score_from_turn()
        return self.overall_score
        
        
```

## Initiate Gameplay

```python
player = Player()
game = Game(player)


for _ in range(7):
    game.play_game(player)
    print(game.turns, player.turn_total, game.overall_score)

```

```python

def run_simulation_for_default_player():
    """Allows player to play 7 turns"""
    player = Player()
    game = Game(player)
    final_score = game.play_game(player)
#     print(game.turns, player.turn_total, game.overall_score)
    return final_score   
        
run_simulation_for_default_player()       
```

```python
class SafetyPlayer(Player):
    """Will play until turn_total reaches 15"""

    def play_and_get_turn_total(self):
        self.turn_total = 0
        while self.turn_total <= 15:
            self.points = self.roll()
            if self.points > 1:
                self.turn_total += self.points
            elif self.roll == 1:
                self.turn_total = 0    
        return self.turn_total
```

```python
def run_simulation_for_SafetyPlayer():
    player = SafetyPlayer()
    game = Game(player)
    rolls = []
    for i in range(7):
        rolls.append(player.play_and_get_turn_total())
    return sum(rolls)


run_simulation_for_SafetyPlayer()
```

Generate n Trials for Default Player

```python
def run_simulation_for_trials(n=100):
    results = [run_simulation_for_default_player() for _ in range(n)]
    n += 1
    return results

run_simulation_for_trials(1000)
```

```python
default_player_results = run_simulation_for_trials(1000)
```
