```python
import random
import numpy as np
import statistics as st
```

```python
def cust_time(*args):
    if random.random() > 0.7:
        return random.random() * 3 + 7
    else:
        return random.random() + 2
```

```python
cust_time()
```

```python
def customers(size=100):
    return [cust_time() for _ in range(size)]
```

```python
customers(100)
```

```python
def trial_data(trials=1000):
    return np.array([customers() for _ in range(trials)])
```

```python
trials = trial_data()
```

```python
trials.shape
```

```python
trials.sum(axis=1)
```

```python
sums = trials.sum(axis=1)
```

```python
mean_people_minutes = sums.mean()
```

```python
mean_people_minutes / 60
```

```python
(mean_people_minutes + sums.std()) / 60
```
